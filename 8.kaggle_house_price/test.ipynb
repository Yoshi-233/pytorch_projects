```python
import secrets
import string
import random

# 指定字符串的长度
length = random.randint(100, 1000)

# 生成随机字符串
random_string = ''.join(secrets.choice(string.ascii_letters + string.digits) for _ in range(length))
print(random_string)
print(len(random_string))
```
